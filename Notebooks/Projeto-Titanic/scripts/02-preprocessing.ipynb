```python
# load base de treinamento
data = pd.read_csv('./../datasets/raw/train.csv')
data.head()
```

```python
data.sample(10)
```

```python
# missing values

""" 
So the first step is to see if your data has missing values. If your data has missing values, you need to check the distribution of each numerical variable (with missing values). If the values in the numerical variables are missing, use the Mean value if the variable is in a normal distribution. Otherwise, choose Median. And if the variable is categorical or discrete, you can select mode. So you need to choose a different measure for each variable. I hope you liked this article on how to fill in missing values in your data. Feel free to ask valuable questions in the comments section below.
"""

data.isna().sum()
```

```python
# AGE: estat√©gia para maior acertividade. dividir em subgrupos e colocar a mediana de cada um deles

temp = data.groupby(['Sex', 'Pclass', 'Survived'])['Age'].median()

for row in zip(temp.index, temp.values):
    indexes = data[\
            (data['Age'].isna()) \
            & (data['Sex'] == row[0][0]) \
            & (data['Pclass'] == row[0][1]) \
            & (data['Survived'] == row[0][2])].index
    
    data.loc[indexes, ['Age']] = row[1]
#     display(row, data.loc[indexes, ['Age']])
```

```python
data.isna().sum()
```

```python
for pclass in data['Pclass'].unique():
    for survived in data['Survived'].unique():
        mode = data[\
            (data['Cabin'].notna()) \
                & (data['Pclass'] == pclass) \
                & (data['Survived'] == survived)]['Cabin'].mode()[0]
        indexes = data[\
                    (data['Cabin'].isna()) \
                        & (data['Pclass'] == pclass) \
                        & (data['Survived'] == survived)].index
        
        data.loc[indexes, ['Cabin']] = mode

        print(pclass, survived, mode, indexes, data.loc[indexes, ['Cabin']])
        # display(data[(data['Cabin'].isna()) & (data['Pclass'] == pclass) & (data['Survived'] == survived)])
```
