TITANIC DATASET - EDA
=====================

```python
import pandas as pd 

from matplotlib import pyplot as plt 
import seaborn as sns 

sns.set_style('darkgrid')
```

```python
data = pd.read_parquet('./../datasets/silver/train-not-null.parquet')
```

```python
display(data.shape, data.sample(10))
```

```python
data.describe()
```

## Plots

```python
sns.barplot(data['Survived'].value_counts(normalize=True))
plt.title("Contagem da classe (target) em PCT");
```

```python
sns.histplot(data['Age'], kde=True)
plt.title("Distribuio por idade");
```

```python
sns.countplot(data,x='Pclass', hue='Sex')
plt.title("Distribuio por classe de embarque e gnero");
```

```python
sns.countplot(data,x='Pclass', hue='Survived')
plt.title("Distribuio por classe de embarque e taxa de sobrevivncia\n0 = No, 1 = Yes");
```

```python
sns.countplot(data, x='Embarked', hue='Pclass')
plt.title('Distribuio de classe de embarque e\n porto de embarque de cada passageiro');
```

## Correlaes

```python
correlation_matrix = data.corr(numeric_only=True)
correlation_matrix.style.background_gradient(cmap='viridis')
```

```python
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='viridis');
```
